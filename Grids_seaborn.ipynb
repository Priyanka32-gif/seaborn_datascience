```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
#seeing the unique value of species
```

#pairplot() is the automated joint plot
# it will automatically create pair plot for us

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

# we use grid plot to have control over dataset graph

```python
# to make a scatter plot
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
#for diagonal,upper and lower plot
g.map_diag(sns.displot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# in FacetGrid we can specify column and row of grid map
g= sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
# seperationg the result based on column and row  on certain colimn
g= sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot,'total_bill')
```

```python
#when we need two argument
g= sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot,'total_bill','tip')
```

```python
g= sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

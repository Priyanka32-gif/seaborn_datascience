Regression plot
l and m plot
it will allow to display linear models with seaborn
how to change size and aspect ratio for seaborn plot

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
#it is fast grid call 
#replacing hue argument with col and row 

sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',
           hue='sex')
#for more info
```

```python
#changing aspect ratio
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue=
          'sex', aspect=0.6)
```

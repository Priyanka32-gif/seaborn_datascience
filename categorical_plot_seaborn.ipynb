```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

# Categorical Plots

```python
sns.barplot(x='sex', y='total_bill',data=tips)
```

```python
sns.barplot(x='sex', y='total_bill',data=tips,estimator=np.std)
#estimator show the estimation of catagorized value with line 
```

```python
sns.countplot(x='sex',data=tips)
#count() function count the number of occurances
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
#boxplot shows the core tiles of the data set while the whiskers extend 
# shows the rest of the distribution 
# dots are outliers
```

```python
# hue argument allow us to splot up these box plots further by another
# categorical column
sns.boxplot(x='day',y='total_bill',data=tips, hue= 'smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# violin plot also show the data on some sort of category 
sns.violinplot(x='day',y='total_bill',data=tips)
```

voilen plot allow us to actually plot all the components that correspond
to actual data points and it's essentially shows the kernel density 
estimation of the underlying distribution

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# strip plot draw a scatterplot where one variable is categorical
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
# jitter argument seperate some of these stacked points
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips)
# its better to do swarm plot with violin plot
```

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') #loading dataset
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill']) # calling distribution plot function
#use bins for changing plot details
```

```python
#joinplot combines two displot and helps to compare two variable
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
#kind key holds to value that affect the data representation
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
#kde represent the density of data
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#pairplot shows the pair relations like:
sns.pairplot(tips)
```

```python
# hue argument show data catagorizing based on given column
sns.pairplot(tips,hue='sex')
```

```python
# rugplot argument draws mark for every points on this uniforn 
# or unique variant distribution 
sns.rugplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'],kde=False)
```

```python
sns.kdeplot(tips['total_bill']) # kde => kernal density estimation
```

```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
tips.head()
```

```python

sns.countplot(x='sex',data=tips)
```

```python
#set_style helps to set the style for rest of the plot
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
#set_style helps to set the style for rest of the plot
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
#set_style helps to set the style for rest of the plot
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
#set_style helps to set the style for rest of the plot
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
#setting size 
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
#setting context
sns.set_context('poster')
sns.countplot(x='sex',data=tips)
```

```python
#setting fonts
sns.set_context('poster',font_scale=3)
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('notebook')
sns.countplot(x='sex',data=tips)
```

```python
#pallet argument change the color of the plot
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',
           palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',
           palette='seismic')
```
